```python
import mysql.connector # DB 연결
from konlpy.tag import Okt #형태소 분석
from collections import Counter#명사카운터  
import pandas as pd 
from wordcloud import WordCloud
import seaborn as sns #막대그래프
import matplotlib.pyplot as plt # 시각화 라이브러리 
from sqlalchemy import create_engine #DB(muddbase2) 연결
from sqlalchemy.orm import sessionmaker #DB(muddbase2) 연결
```

```python
# DB(muddbase) 연결
local = mysql.connector.connect (
     host = 'localhost',
     user = 'root',
     password = "1234",
     database = "muddbase"  
)
cur = local.cursor()  #커서생성
```

```python
# tb_review_detai, tb_answered_keyword에서 comment 컬럼 데이터 가져오기
cur.execute("select CONCAT(trd.item_comment, tak.related_keyword_comment) FROM tb_review_detail trd join tb_answered_keyword tak on trd.mb_email = tak.mb_email")
text_sum = cur.fetchall()
text_sum
```

```python
# 형태소 분석 및 단어 빈도 계산하는 함수

def calculate_word_frequency(text_sum):
    # 형태소 분석기 초기화 (의미있는 정보 추출: 빈도분석, 품사, 의미, 언어이해)
    okt = Okt()
    
    # 모든 토큰을 저장할 리스트
    all_tokens = []

    # 사원 이름의 데이터 불러와서 익명처리를 위한 이름 추출 
    cur.execute ("SELECT mb_name from tb_member")
    remove_name= cur.fetchall()
    remove_name = [name[0] for name in remove_name]  # 튜플에서 이름만 추출

    # 이름 리스트에서 제거할 토큰 목록 만들기
    # set() : 중복을 허용하지 않는 리스트 
    remove_tokens = set()
    for name in remove_name:
        # 이름의 모든 부분 문자열 추가 ( "홍", "길", "동", "홍길", "길동" 등 모든 조합이 포함)
        length = len(name)
        for i in range(length):
            for j in range(i + 1, length + 1):
                remove_tokens.add(name[i:j])
                
    # 각 텍스트에서 토큰화 및 단어 빈도 계산
    for row in text_sum:
        text = row[0]   # 튜플의 첫번째 요소에서 텍스트 추출
        tokens = okt.nouns(text) # 형태소 분석 수행 후 명사만 추출
        
        # remove_name 리스트 속의 단어를 제외한 단어만 남김 (2단어 이상)
        filtered_tokens = [token for token in tokens if token not in remove_tokens and len(token) > 1]  # text 객체 속의 단어를 제외한 단어만 남김
        
        all_tokens.extend(filtered_tokens)

    # 단어 빈도 계산
    word_freq = Counter(all_tokens)   
    
    return word_freq

# 확인용 빈도 계산
word_freq = calculate_word_frequency(text_sum)

# 데이터프레임 생성
df_word_freq = pd.DataFrame(word_freq.items(), columns=['Word', 'Count'])

# Count 컬럼을 기준으로 내림차순 정렬
df_word_freq = df_word_freq.sort_values(by='Count', ascending=False)

# 상위 30개 단어 추출
top30 = df_word_freq.head(30)

# 결과 출력
print(top30)
```

```python
# 막대그래프
plt.rcParams.update ({ 'font.family' : 'Malgun Gothic',
                      'figure.dpi' :'120',
                      'figure.figsize':[6.5,6]
})
sns.barplot(data= top30, y= 'Word', x='Count')
```

```python
# 워드 클라우드 만들기 (상위 30개)
dic_word = top30.set_index('Word').to_dict()['Count']

dic_word

font = '‪C:\Windows\Fonts\gulim.ttc'

wc = WordCloud( random_state = 100, 
                font_path = font,
                width = 400,
                height = 400,
                background_color = 'white'
              ) 

img_wordcloud = wc.generate_from_frequencies(dic_word)
plt.figure(figsize = (5,5))
plt.axis('off')  
plt.imshow (img_wordcloud)

# 워드 클라우드 이미지를 파일로 저장
img_wordcloud.to_file('wordcloud1.png')

# 저장된 이미지 파일을 base64 형식으로 읽어옴
import base64
with open('wordcloud.png', 'rb') as f:
    img_data = f.read()

# base64로 인코딩하여 JSON 형식으로 변환
img_wordcloud_json = base64.b64encode(img_data).decode('utf-8')
```

```python
# 분석 결과 데이터 DB(muddbase2)연결
engine = create_engine('mysql+mysqlconnector://root:1234@localhost/muddbase')
Session = sessionmaker(bind=engine)
session = Session()

try:
    # 이 부분을 본인의 설정에 맞게 변경 (if_exists='replace'는 기존 테이블을 대체, 'append'는 데이터 추가)
    top30.to_sql(name='tb_textmining', con=engine, if_exists='replace', index=True)
    session.commit()
    print("DataFrame successfully written to database")
except Exception as e:
    print(f"Error: {e}")
finally:
    session.close()

cur.close()
local.close()
```
